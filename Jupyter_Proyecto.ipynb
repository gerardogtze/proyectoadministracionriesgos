```python
# Ejemplo AAPL MSFT cambiar por activos correctos
data = yf.download("XOM ALV.DE BOOT.L CVX DGE.L JNJ.DE", start="2021-01-01", end= "2022-12-31")

data = data.dropna()

rends = ((data['Adj Close'] / data['Adj Close'].shift(1) - 1).fillna(0))*100

data['Rendimiento XOM']= rends["XOM"]
data['Rendimiento XOM']=data['Rendimiento XOM'].round(2)
data['Rendimiento XOM']=data['Rendimiento XOM'].astype(str)+ '%'

data['Rendimiento ALV.DE']= rends["ALV.DE"]
data['Rendimiento ALV.DE']=data['Rendimiento ALV.DE'].round(2)
data['Rendimiento ALV.DE']=data['Rendimiento ALV.DE'].astype(str)+ '%'

data['Rendimiento BOOT.L']= rends["BOOT.L"]
data['Rendimiento BOOT.L']=data['Rendimiento BOOT.L'].round(2)
data['Rendimiento BOOT.L']=data['Rendimiento BOOT.L'].astype(str)+ '%'

data['Rendimiento CVX']= rends["CVX"]
data['Rendimiento CVX']=data['Rendimiento CVX'].round(2)
data['Rendimiento CVX']=data['Rendimiento CVX'].astype(str)+ '%'

data['Rendimiento DGE.L']= rends["DGE.L"]
data['Rendimiento DGE.L']=data['Rendimiento DGE.L'].round(2)
data['Rendimiento DGE.L']=data['Rendimiento DGE.L'].astype(str)+ '%'

data['Rendimiento JNJ.DE']= rends["JNJ.DE"]
data['Rendimiento JNJ.DE']=data['Rendimiento JNJ.DE'].round(2)
data['Rendimiento JNJ.DE']=data['Rendimiento JNJ.DE'].astype(str)+ '%'



data.drop(columns=["Open", "Close", "Low", "High"])
```

```python
rendimientos_portafolio= rends.mean()
volatilidad_portafolio= rends.std().mean()
rf= 0.0338 
sharpe_ratios = (rendimientos_portafolio - rf) / volatilidad_portafolio
sharpe_ratios
```

```python
# Calcular los rendimientos diarios de los activos
returns = data["Adj Close"].pct_change()

# Definir la función de utilidad para calcular los retornos, la varianza y el Sharpe Ratio del portafolio
def calcular_rendimiento_portafolio(weights, returns):
    portfolio_return = np.sum(returns.mean() * weights) * 252
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
    sharpe_ratio = portfolio_return / portfolio_volatility
    return np.array([portfolio_return, portfolio_volatility, sharpe_ratio])

# Definir la función de optimización para encontrar el portafolio con el Sharpe Ratio más alto
def optimizacion_sharpe_ratio(returns):
    num_activos = len(returns.columns)
    pesos_iniciales = np.array(num_activos * [1.0 / num_activos])

    # Restricciones de pesos (weights)
    restricciones = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1},
              {'type': 'ineq', 'fun': lambda x: 0.2 - x[0]},
              {'type': 'ineq', 'fun': lambda x: 0.2 - x[1]},
              {'type': 'ineq', 'fun': lambda x: 0.2 - x[2]},
              {'type': 'ineq', 'fun': lambda x: 0.2 - x[3]},
              {'type': 'ineq', 'fun': lambda x: 0.2 - x[4]},
              {'type': 'ineq', 'fun': lambda x: 0.2 - x[5]})

    # Limites de pesos (weights)
    limites_pesos = tuple((0, 1) for x in range(num_activos))

    # Optimización de Sharpe Ratio
    resultado = sco.minimize(fun=lambda x: -calcular_rendimiento_portafolio(x, returns)[2],
                         x0=pesos_iniciales,
                         method='SLSQP',
                         bounds=limites_pesos,
                         constraints=restricciones)
    return resultado
# Llamar la función de optimización para obtener el portafolio de mayor Sharpe Ratio
opt = optimizacion_sharpe_ratio(returns)['x']
opt_names = {
    'XOM':opt[0],
    'ALV.DE':opt[1],
    'BOOT.L':opt[2],
    '0R1G.IL':opt[3],
    'DGE.L':opt[4],
    'JNJ.DE':opt[5],
}
opt_names
```
